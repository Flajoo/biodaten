## 3. Repeat: Quality control
### 3.1 Overview
Use wordcount (```wc```) to check how many reads are in each of the both fastq files!

`````bash
wc -l ./BC1_notrim.fastq
`````

`````bash
head -6 ./BC1_notrim.fastq #show only the 6 first rows of a file 
`````

`````bash
# You can also use the Stream Editor (SED) to print certain lines to your command line.

sed -n '1,6p' ./BC1_notrim.fastq 

# in this example we use -n to suppress automatic printing of all lines
# then we select the row 1 to 6 
# 'p' prints the selected row 
`````

A Fastq file always consists of four lines. The first line (allways starts with '@')is the annotation line of the read, followed by the line containing the actual nucleotide sequence. The third line contains a ‘+’ and the fourth line contains the quality of each nucleotide from line two.

The easiest way to get the correct number of reads would be to divide the output from ```wc``` by 4 but you can also use SED again to count only every 4th line

`````bash
sed -n '1~4p' ./BC1_notrim.fastq | wc -l

# 1~4 selects only every 4th line starting from the first line
`````

`````bash
sed -n '1,40{1~4p}' ./BC1_notrim.fastq

# this can also be combined to get a list of the first 10 annotation lines.
`````
